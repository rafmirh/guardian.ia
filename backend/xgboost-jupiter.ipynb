```python
import pandas as pd
```

```python
personasf_df = pd.read_csv('persona_fisica.csv')
personasf_df.head()
```

```python
personasf_df['fecha_inicio'] = pd.to_datetime(personasf_df['fecha_inicio'])
personasf_df['anio_inicio'] = personasf_df['fecha_inicio'].dt.year
personasf_df['mes_inicio'] = personasf_df['fecha_inicio'].dt.month
personasf_df['dia_inicio'] = personasf_df['fecha_inicio'].dt.day

personasf_df['fecha_hecho'] = pd.to_datetime(personasf_df['fecha_hecho'])
personasf_df['anio_hecho'] = personasf_df['fecha_hecho'].dt.year
personasf_df['mes_hecho'] = personasf_df['fecha_hecho'].dt.month
personasf_df['dia_hecho'] = personasf_df['fecha_hecho'].dt.day

personasf_df
```
