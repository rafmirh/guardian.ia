```python
# Importamos la librería pandas
import pandas as pd

# Leemos el archivo csv con pd.read
df = pd.read_csv('https://archivo.datos.cdmx.gob.mx/FGJ/victimas/victimasFGJ_2024.csv')
df.head()
```

```python
# Valores únicos de la columna delito
df['delito'].unique()
```

```python
# Filtración del delito que usaremos (usurpación de identidad)
robo_identidad = df[df['delito'] == 'USURPACIÓN DE IDENTIDAD']
robo_identidad.head()
```

```python
# Información de la filtración
robo_identidad.info()
```

```python
# Conteo de valores de la columna tipo_persona
robo_identidad['tipo_persona'].value_counts()
```

```python
# Filtración personas MORAL
persona_moral = robo_identidad[robo_identidad['tipo_persona']== 'MORAL']
```

```python
# Eliminación de las columnas sexo y edad
persona_moral = persona_moral.drop(columns=['sexo', 'edad'])
```

```python
# Eliminación de valores NAN en persona MORAL
persona_moral.dropna()
```

```python
# Filtración de persona FISICA
persona_fisica = robo_identidad[robo_identidad['tipo_persona']== 'FISICA']
```

```python
# Eliminación de valores NAN en persona FISICA
persona_fisica.dropna()
```

```python
# Exportación de archivo csv de persona FISICA
persona_fisica.to_csv('persona_fisica.csv', index=False)
```

```python
# Exportación de archivo csv de persona MORAL
persona_moral.to_csv('persona_moral.csv', index=False)
```
